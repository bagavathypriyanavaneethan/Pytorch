**y=x^2**

```python
x=torch.tensor(4.0,requires_grad=True)   #To execute back propogation too
```

```python
x
```

```python
y=x**2
```

# Backpropogation 2x

```python
y.backward()
```

```python
x.grad
```

# Complex tensor

```python
lst=[[2.,3.,1.],[4.,5.,3.],[7.,6.,4.]]

torch_input=torch.tensor(lst,requires_grad=True)
```

```python
torch_input
```

**y=x^3+x^2**

```python
y=torch_input**3 + torch_input**2
```

```python
y
```

```python
z=y.sum()
```

```python
z
```

```python
z.backward()
```
