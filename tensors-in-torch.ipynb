# Using torch.tensor

```python
ten_arr=torch.tensor(arr)
```

**It creates a another copy of arr to ten_arr**

```python
ten_arr[3]=50

print(ten_arr)

print(arr)
```

```python
torch.zeros(2,3,dtype=torch.float32)
```

# Arithmetic operation

```python
a=torch.tensor([1,2,3],dtype=torch.float32)

b=torch.tensor([5,6,7],dtype=torch.float32)

print(a+b)
```

```python
torch.add(a,b)
```

```python
c=torch.zeros(3)
```

**Storing in a variable**

```python
torch.add(a,b,out=c)
```

```python
c
```

```python
torch.add(a,b).sum()
```

# Dot product and multiplication operation

**Multiplication**

```python
print(a)

print(b)

a.mul(b)
```

```python
a.dot(b)       #1.5 + 2.6 + 3.7
```

# Matrix multiplicaiton

```python
a=torch.tensor([[4,5,6],[7,2,8]],dtype=torch.float)

b=torch.tensor([[3,2,1],[7,1,5],[1,2,3]],dtype=torch.float)

torch.matmul(a,b)
```

```python
a@b
```

```python
torch.mm(a,b)
```
